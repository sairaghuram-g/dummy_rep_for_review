```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
z = [1, 2, 3]
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num/den

print(sigma)
print(np.sum(sigma))
```

```python
z = np.random.randint(-5, high=15, size=25)
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num/den

print(sigma)
print(np.sum(sigma))
```

```python
plt.plot(z, sigma, 'ko')
plt.xlabel('z')
plt.ylabel('softmax')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
plt.plot(z, sigma, 'ko')
plt.xlabel('z')
plt.ylabel('softmax')
plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

If we see, a softmax function is basically a linear regression function on a logarithmic scale
